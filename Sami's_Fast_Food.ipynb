###**Problem Statment:**
#####Using dictionaries, create a program that can do the job of a cashier in a fast-food restaurant.

###**Expected Final Output:**
#####The final outcome which I expect from a user who orders 3 items should look something like this:

##### "User here is your final order of ['Pizza', 'Onion Rings', 'Sprite',] , costing you $11.45 enjoy your day and come back again!"

###**Logic used to solve the problem:**
#####Using dictionaries I stored all the dishes in the same category in the same dictionary and I had an empty list in which I stored the user's order. I used the name of the dish as the key and the value of the dish as the price for that dish. I would take the user's input after I asked them which meal that they would select and then first see if they would like that meal by putting it into an if statement and seeing if their input was no. If they did order a meal the program goes through another if statement which searches to see if the user's input was in the dictionary key for that meal. If it wasn't in the dictionary keys the program ends. If the input is in the dictionary keys the program takes the user's input to see what dictionary key matches it then it gets a value of that key which is turned into a variable to save the price of what they ordered. This is repeated 4 times to get the user through all the categories of meals. Then the program adds all the prices of the meals together by adding all the price variables together. Using an if statement the program checks if the total price is equal to 0. This used to make sure that the user didn't order anything as if the user ordered something the total price would have value. If the total price is equal to 0 then the program ends, but if it has a value then the program uses an if statement to confirm that the user wants to order the meal. If the user says yes the program uses time.sleep and a for loop which prints "..." to pretend that the food is being cooked. Finally, the program prints the list where all the orders are saved and the total price.

```python
#Importing modules
import sys
import time

#dictionaries for all dishes on menu
main_dish = {"Burger" : 6.45, "Pizza":7.25, "Taco":5.75,"Hot Dog":5.45}
side_dish = {"Fries": 2.45, "Onion Rings": 2.75, "Chicken Nuggets":4.45, "Salad":2.25}
drink = {"Coke": 1.75, "Sprite": 1.45, "Water":0.25, "Coffee":2.25}
dessert = {"Ice Cream":3.45, "Brownies":3.25, "Chocolate Cookies":2.45}
#empty list for orders to store the users order

order = []


#printing list for main dishes
name = input("Welcome to Sami's Fast Food what is your name? \n" )
time.sleep(1.5)
print("Hey %s here are our main dishes" %name)
print(main_dish)

time.sleep(1.5)
main = input("Which main meal will you select? \n (Type 'N' if you don't want a main meal) \n" )

time.sleep(1)
#searching if user input matches what is in the main_dish list
if main.upper() == "N":
  print("Ok so you are not ordering a main meal")
  p1 = 0
else:

  if main in main_dish.keys():
    
    p1 = main_dish[main]
    print(main, "has been added to your order!")
    #adding users input to the orders list
    order.append(main)
  else:
    print("Sorry but", main, "is not one our main dishes, please try ordering again!")
    #ending program if users input is invalid
    sys.exit()





#Side dish
time.sleep(0.5)
print("For our side dishes we are serving \n", side_dish)
time.sleep(1.5)

side = input("Please select your side dish \n (Type 'N' if you don't want a side dish) \n")

if side.upper() == "N":
  print("Ok so you are not ordering a side dish")
  p2 = 0

else:
  time.sleep(1)
  if side in side_dish.keys():

    p2 = side_dish[side]
    print(side, "have been added to your order!")
    order.append(side)
  else:
    print("Sorry but", side, "is not one of our side dishes, please try ordering again!")
    sys.exit()




#Drinks
time.sleep(0.5)
print("Our drinks consist of \n", drink)
time.sleep(1.5)
drinks = input("Which drink will you select? \n (Type 'N' if you don't want a drink)  \n")

time.sleep(1)
if drinks.upper() == "N":
  print("Ok so you are not ordering a drink")
  p3 = 0
  
else:
  if drinks in drink.keys():
    p3 = drink[drinks]
    
    print(drinks, "has been added to your order!")
    order.append(drinks)
  else:
    print("Sorry but", drinks , "is not one of our drinks, please try ordering again!")
    sys.exit()




#Dessert
time.sleep(0.5)
print("Finally, for dessert we have \n", dessert)
time.sleep(1.5)
desserts = input("Which dessert will you select? \n (Type 'N' if you don't want a dessert) \n")

time.sleep(1)
if desserts == "N":
  print("Ok so you are not ordering a dessert")
  p4 = 0

else:

  if desserts in dessert.keys():
    p4 = dessert[desserts]
    print(desserts, "has been added to your order!")
    order.append(desserts)
  else:
    print("Sorry but", desserts , "is not one of our desserts, please try ordering again!")
    sys.exit()

#Adding all costs to get total price
p5 = p1+p2+p3+p4
#Formatting the final cost to have 2 decimal places.
totp = (format(p5, '.2f'))

#Making sure user didn't order nothing
if totp == 0:
  print("Sorry but you can't order nothing. \n Try ordering again.")

else:

  #Confirming if user wants to proceed with order
  Proc = input("Your total order will cost $%s \n Would you still like to proceed with your order? \n Y/N \n" %totp)

  time.sleep(0.5)

  if Proc.upper() == "Y":
    print("Please wait while your order is prepared %s!" %name)

    #Using for loop to print "..." 3 times while preparing food
    for i in range(3):
      print(".")
      time.sleep(1)
      print("..")
      time.sleep(1)
      print("...")
      time.sleep(1)

    #Print users order from the order list
    print(name, "here is your final order of", order, ", costing you $%s enjoy your meal and come back again!" %totp)
  else:
    print(name, "if you would like to order again please press play again.")
```
